```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars = star_wars[star_wars['RespondentID'].notnull()]
v = star_wars[star_wars.columns[1]].value_counts()
print(v)
star_wars.head(10)

```

```python

boo = {"Yes":True, "No":False}
star_wars[star_wars.columns[1]] = star_wars[star_wars.columns[1]].map(boo)
star_wars[star_wars.columns[2]] = star_wars[star_wars.columns[2]].map(boo)

bools = {"Star Wars: Episode I  The Phantom Menace":True,
         "Star Wars: Episode II  Attack of the Clones":True,
         "Star Wars: Episode III  Revenge of the Sith":True,
         "Star Wars: Episode IV  A New Hope":True,
         "Star Wars: Episode V The Empire Strikes Back":True,
         "Star Wars: Episode VI Return of the Jedi":True,
         np.nan: False}

for c in range(3,9):
    star_wars[star_wars.columns[c]] = star_wars[star_wars.columns[c]].map(bools)

star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1",
                                     "Unnamed: 4": "seen_2", "Unnamed: 5": "seen_3", "Unnamed: 6":"seen_4",
                                     "Unnamed: 7":"seen_5", "Unnamed: 8": "seen_6"})
star_wars.head(10)
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1",
                                      "Unnamed: 10":"ranking_2", "Unnamed: 11": "ranking_3",
                                     "Unnamed: 12":"ranking_4", "Unnamed: 13": "ranking_5",
                                     "Unnamed: 14": "ranking_6"})

means = []
movies = list(star_wars.columns[9:15].values)
for item in range(9,15):
    means.append(star_wars[star_wars.columns[item]].mean())

print(means)
print(movies)
```

```python
views = ['seen_1','seen_2','seen_3','seen_4','seen_5','seen_6']
viewed = star_wars[views].sum(axis=0)

fig = plt.figure(figsize=(20,6))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

N = 6
ind = np.arange(N)
width = 0.70

bar1 = ax1.bar(ind, viewed, color = 'b')
ax1.set_xticklabels(views)
ax1.set_ylabel('# of Views')
ax1.set_title('Most Viewed Star Wars Films')
ax1.set_xticks(ind + 0.5*width)

bar2 = ax2.bar(ind, means, width, color = 'r')
ax2.set_xticklabels(movies)
ax2.set_ylabel('Avg Rank')
ax2.set_title('Average Ranking of Star Wars Films')
ax2.set_xticks(ind + 0.5*width)

plt.show()
```

```python
#Views seem correlated with the ranking#
corr = np.corrcoef(means, viewed)
print(corr)
```

## Why are the originals better? ##
- George Lucas had limited budget and huge number of rewrites for the original. Was constrained by the studio and wasn't a living legend at the time. Supposedly Harrison Ford gave pushback on dialogue which is why Han Solo comes across as the most normal character 
- For the prequels Lucas was given free reign, and he just isn't that good of a director. Over reliance on CGI which was all the rage for the late 90's
- Episode 7 was pretty good which used the standard Disney formula, but also used real actors and sets which gives the movie a much better feel

## Generational Gap ##
- Views are highly correlated to ranking except for episode 1
- People saw episode 5 because of the hype, and it really was the best one
- As a result people saw episode 1 because of expectations, but it was so bad that they gave up on watching the rest of the prequels 

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

ranks = ['ranking_1','ranking_2','ranking_3','ranking_4','ranking_5','ranking_6']
views = ['seen_1','seen_2','seen_3','seen_4','seen_5','seen_6']

male_viewed = males[views].sum(axis=0)
male_average = males[ranks].mean(axis=0)

female_viewed = females[views].sum(axis=0)
female_average = females[ranks].mean(axis=0)

N=6
width=0.35
ind = np.arange(N)

fig = plt.figure(figsize=(20,6))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

m1 = ax1.bar(ind, male_viewed, width, color = 'b')
w1 = ax1.bar(ind + width, female_viewed, width, color = 'y')

m2 = ax2.bar(ind, male_average, width, color = 'b')
w2 = ax2.bar(ind + width, female_average, width, color = 'y')

ax1.set_ylabel('Views')
ax1.set_title('Male and Female Raw Views')
ax1.set_xticks(ind + width)
ax1.set_xticklabels((views))
ax1.legend((m1[0], w1[0]), ('Men', 'Women'))

ax2.set_ylabel('Avg Rank')
ax2.set_title('Male and Female Avg Rank')
ax2.set_xticks(ind + width)
ax2.set_xticklabels((ranks))
ax2.legend((m1[0], w1[0]), ('Men', 'Women'))

plt.show()
```

- There are more male respondents to the survey, but comparing the results by gender the same trend is mostly observed with the exception of episode 1 and episode 4 where females seem to rate episode one more favourably

```python
print(male_average)
print(female_average)
```

```python
fig = plt.figure(figsize=(20,6))
ax = fig.add_subplot(111)

ax.scatter(male_viewed, male_average, s=100, c='b', marker="o", label='Males')
ax.scatter(female_viewed,female_average, s=100, c='r', marker="o", label='Females')
ax.set_ylabel('Average Rank')
ax.set_title('Scatterplot of Views vs Avg Rank by Gender')
ax.set_xlabel('# of Views')
plt.legend(loc='upper right');

labels = ['Episode {0}'.format(i) for i in range(1,7)]
for label, x, y in zip(labels, male_viewed, male_average):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (50, 20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'grey', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
for label, x, y in zip(labels, female_viewed, female_average):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (-50, -20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'grey', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))

plt.show()
```

## Discussion ##
- The female respondents actually follow the more linear trend, with more views always leading to a better ranking
- For the males, episode 4 is the outlier in that it has a high ranking but fewer views
- The interesting point is that of all respondents, episodes 5 and 6 are the most watched which are part of the original trilogy, but more people have viewed episode 1 than episode 4. So there is some obvious disconnect in the order that most people have watched the movies

```python
#look at female respondents who have viewed both episode 1 and 4#
female_1_4 = females[(females['seen_1'] == True) & (females['seen_4'] == True)]
female_average_1_4 = female_1_4[ranks].mean(axis=0)
female_viewed_1_4 = female_1_4[views].sum(axis=0)
print(female_1_4.shape[0])
print(female_average_1_4)
```

## Relief ##
- When looking at female respondents who have seen both episodes 1 and 4, the same pattern emerges as the male respondents
- The reason for the unusually high ranking of episode 1 amongst women is possibly caused by skewed data resulting from individuals who have seen the prequels but not the original trilogy

```python
N=6
width=0.35
ind = np.arange(N)

fig = plt.figure(figsize=(20,6))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

m1 = ax1.bar(ind, female_viewed_1_4, width, color = 'b')
w1 = ax1.bar(ind + width, female_viewed, width, color = 'y')

m2 = ax2.bar(ind, female_average_1_4, width, color = 'b')
w2 = ax2.bar(ind + width, female_average, width, color = 'y')

ax1.set_ylabel('Views')
ax1.set_title('1+4 Sample vs All Women')
ax1.set_xticks(ind + width)
ax1.set_xticklabels((views))
ax1.legend((m1[0], w1[0]), ('1+4', 'Women'))

ax2.set_ylabel('Avg Rank')
ax2.set_title('1+4 Females vs Female Avg Rank')
ax2.set_xticks(ind + width)
ax2.set_xticklabels((ranks))
ax2.legend((m1[0], w1[0]), ('1+4', 'Women'))

plt.show()
```
